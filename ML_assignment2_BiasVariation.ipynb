## 기본함수 정의

```python
import numpy as np
import math  #library 호출
import matplotlib.pyplot as plt
from numpy.linalg import norm
```

```python
def true_f(x):
    return np.sin(x) #true function 정의
```

```python
def gen_pi_data(x, degree): 
    X=np.zeros((x.shape[0],degree+1))
    for d in range (degree+1):
         X[ :, d] = x ** d
    return X # degree에 따른 pi data set 생성
```

```python
def fit_poly(X,y,w_array,y_array, y_array_2):
    Y = y.reshape(-1,1)
    A = np.matmul(X.transpose(), X)
    B = np.matmul(X.transpose(), Y)
    weight = np.linalg.solve(A,B)
    y_pred = X @ weight
    
    w_array.append(weight) # 계산한 weight 값을 weight array에 추가
    y_array.append(y_pred) # model로 추정한 y값을 y array에 추가
    y_array_2.append(y_pred)
   
```

## Sample 생성

```python
num_sample = 21 # sample의 개수

x_train = np.linspace(-math.pi, math.pi, num_sample) # Xdata 생성
#true_f(x) = np.sin(x) # True Function 


```

# Finding Expectation of models
    

```python
degree = 9 # polynomial degree의 최대값
num_experiment = 100 #실험 횟수
avg_W_set=[] # 실험을 통한 model 평균값 배열 , 원소개수 degree+1
weight=[] # 평균 weight 계산하기 위한 배열
p_y=[] # 실험마다 y_pred값 저장
Y_total=[] #실험마다의 y_pred 값인데 p_y는 초기화시켜서.
Y_val=[] # 평균 model을 통해 측정한 y_pred값
var_set=[] # degree에 따른 variation vector (표준편차벡터)

V_max=[] # Y_val + 표준편차 set
V_min=[] # Y_val - 표준편차 set
var_set_d=[] # Variation 대표값 set
bias_set=[] # bias 대표값 set


for d1 in range(degree+1):

    avg_w = np.zeros(((d1+1),1))
    
    for n1 in range(num_experiment):
        Y=true_f(x_train) + np.random.normal(loc= 0, scale= 0.5, size= num_sample)# 실험 횟수만큼 data sample 생성
        X_pi = gen_pi_data(x_train, d1)
        fit_poly(X_pi,Y,weight,p_y, Y_total)
        avg_w = (avg_w + weight[n1])

    avg_w = avg_w / (num_experiment)#평균 weight 계산
    avg_W_set.append(avg_w)#degree에 따른 w 평균
    X_val = gen_pi_data(x_train, d1)#평균으로 구한 모델 실험할 데이터
    Y_val.append(X_val @ avg_W_set[d1])#평균 model의 prediction값
    
    
    #=======================Variaition======================
    avg_v = np.zeros((x_train.shape[0],1))
    var = np.zeros((x_train.shape[0],1))
    ex_var=0
    variance=0
    for n2 in range(num_experiment):
        var = (((Y_val[d1] - p_y[n2]) **2) **(1/2)) #표준편차
        avg_v = avg_v + var #표준편차 벡터
        variance += norm(Y_val[d1] - p_y[n2]) # model 마다의 분산합
  
        
    ex_var = variance/ ((num_sample) * (num_experiment)) # 분산 대표값
    avg_v = avg_v / (num_experiment)
    var_set.append(avg_v)
    var_set_d.append(ex_var) # degree에따른 variation set
    
     #=======================Bias============================
    bias = true_f(x_train) - Y_val[d1].reshape(-1) #true - 모델
    ex_bias = (np.matmul(bias, bias.reshape(-1)))/(num_sample)  
    bias_set.append(ex_bias)
     #========================================================
        
    del weight[:] #weight 배열 초기화
    del p_y[:] 

for n1 in range(degree+1):
    V_max.append(Y_val[n1] + var_set[n1])
    V_min.append(Y_val[n1] - var_set[n1])

#그래프 그리기
f, axes = plt.subplots(nrows=2, ncols=5, figsize=(30, 10), dpi=200)

for i in range(0,2):
    for j in range(0,5):
        axes[i,j].plot(x_train, Y_val[(5*i)+j], 'b-')
        axes[i,j].plot(x_train, true_f(x_train), 'r-')
        axes[i,j].set_xlabel('x')
        axes[i,j].set_ylabel('y')
        axes[i,j].set_ylim([-1.1,1.1])
axes[0,0].set_title('Degree=0')
axes[0,1].set_title('Degree=1')
axes[0,2].set_title('Degree=2')
axes[0,3].set_title('Degree=3')
axes[0,4].set_title('Degree=4')
axes[1,0].set_title('Degree=5')
axes[1,1].set_title('Degree=6')
axes[1,2].set_title('Degree=7')
axes[1,3].set_title('Degree=8')
axes[1,4].set_title('Degree=9')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=degree+1, figsize=((degree+1)+8, 1.5), dpi=80)

for m1 in range(degree+1):
    axes[m1].plot(x_train, Y_val[m1])
    axes[m1].set_ylim([-2, 2])
    for m2 in range(num_experiment):
        axes[m1].plot(x_train, Y_total[(m1*100)+m2], 'b-', alpha=0.05)

axes[0].set_title('Degree=0')
axes[1].set_title('Degree=1')
axes[2].set_title('Degree=2')
axes[3].set_title('Degree=3')
axes[4].set_title('Degree=4')
axes[5].set_title('Degree=5')
axes[6].set_title('Degree=6')
axes[7].set_title('Degree=7')
axes[8].set_title('Degree=8')
axes[9].set_title('Degree=9')
```
